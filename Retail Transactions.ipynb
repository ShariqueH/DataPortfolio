```python
#3.Data Cleaning

#Using Z-score Identifying outliers:
```

```python
from scipy import stats
```

```python
#Calculating the Z scores of numeric columns:
z_score_09_10 = stats.zscore(df_09_10.select_dtypes(include=['float64', 'int64']))
z_score_10_11 = stats.zscore(df_10_11.select_dtypes(include=['float64', 'int64']))

#Identifying Outliers:
df_09_10_outliers = df_09_10[(z_score_09_10 > 3). any(axis = 1) | (z_score_09_10 <-3).any(axis=1)]
df_10_11_outliers = df_10_11[(z_score_10_11 > 3). any(axis = 1) | (z_score_10_11 <-3).any(axis=1)]
```

```python
#Removing the Outliers:
df_09_10_no_outliers = df_09_10[(z_score_09_10 < 3). all(axis=1) & (z_score_09_10 > -3).all(axis = 1)]
df_10_11_no_outliers = df_10_11[(z_score_10_11 < 3). all(axis=1) & (z_score_10_11 > -3).all(axis = 1)]
```
