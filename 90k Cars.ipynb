```python
#Preprocessing
```

```python
from sklearn.preprocessing import StandardScaler
import re
from sklearn.model_selection import train_test_split
```

```python
#Converting values into numerical values to make it easy for the computer to understand
```

```python
objects = {col : cars[col].unique() for col in cars.select_dtypes("object")}
numerics = {col for col in cars.columns if cars[col].dtypes != "object"}
df = cars.copy()
for column in objects.keys():
    ordinal = sorted(list(df[column].unique()))
    df[column] = df[column].apply(lambda x : ordinal.index(x))
df.head(5)
```

```python
cars = df
```

```python
#Finding the outliers (as they have a massive impact at the performance of the model)
```

```python
cars.describe()
```

```python
df = cars.copy()
qs = dict()
sums = 0 
for col in numerics:
    wol = df.shape[0]
    
    q1 = np.quantile(df[col],0.25)
    q3 = np.quantile(df[col],0.75)
    iqr = q3-q1
    lower = q1 - (1.4*iqr)
    upper = q3 + (1.4*iqr)
    qs[col + " lowerBound "] = lower 
    qs[col + " Upper Bound  "] = lower

    df = df[(df[col] <= upper) & (df[col] >= lower)]
    df = df.reset_index(drop=True)
    won = df.shape[0]
    qs[col + " outliers "] = wol - won
    sums += wol - won
```

```python
cars = df
```

```python
#Scaling the data to bring it to a unit variance. Seperated target column to not interfere with scaling.
```

```python
input = cars.drop("price",axis=1)
output = cars['price']
```

```python
scaler = StandardScaler()
scaler.fit(input)
input = pd.DataFrame(scaler.transform(input),columns=input.columns)
```

```python
#Scaled
```

```python
input.head()
input.var()
```

```python
#Splitting the data into training data and test data
```

```python
input_train , input_test , output_train , output_test = train_test_split(input,output,random_state=13,train_size=0.745)
```

```python
#Training the model
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
linear_model = LinearRegression()
linear_model.fit(input_train,output_train)
predict = linear_model.predict(input_test)
```

```python
intercept = linear_model.intercept_
coefficients = linear_model.coef_
```

```python
score = mean_squared_error(output_test,predict)
print(score)
```
